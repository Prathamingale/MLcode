```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error , mean_absolute_error ,r2_score 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("weight-height.csv")
print(df.to_string())
```

```python
print(df.shape)
print(df.isnull().sum())
```

```python
x = df.iloc[:,1:2]
y = df.iloc[:,2:3]
```

```python
x
```

```python
y
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size = 0.25,random_state=0)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
ypred = reg.predict(xtest)
```

```python
ypred
```

```python
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
# Visualize the linear regression model
plt.scatter(xtest, ytest, color='blue', label='Actual Data', marker='o', s=1)  # Increase marker size with 's' parameter
plt.plot(xtest, ypred, color='red', linewidth=2, label='Linear Regression Prediction')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Weight-Height Prediction')
plt.legend()
plt.show()
```

```python
# Calculate Mean Squared Error and Mean Absolute Error
mse = mean_squared_error(ytest, ypred)
mae = mean_absolute_error(ytest, ypred)

print("Mean Squared Error =", mse)
print("Mean Absolute Error =", mae)
```

```python
r2 = r2_score(ytest, ypred)
print(f"R-Squared (R2): {r2}")
```
